```python
%matplotlib notebook
import time
import numpy as np
import matplotlib.pyplot as plt

fig,ax = plt.subplots(1,1)

x,y = [], [] 

line, = ax.plot(x, y, 'o')

for i in range(30):
    x.append(np.random.rand())
    y.append(np.random.rand())
    line.set_data(x,y)
    ax.relim()
    ax.autoscale_view(True,True,True)
    fig.canvas.draw()
    time.sleep(0.3)
```
