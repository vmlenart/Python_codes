```python
%%HTML
<script>
  function code_toggle() {
    if (code_shown){
      $('div.input').hide('500');
      $('#toggleButton').val('Show Code')
    } else {
      $('div.input').show('500');
      $('#toggleButton').val('Hide Code')
    }
    code_shown = !code_shown
  }

  $( document ).ready(function(){
    code_shown=false;
    $('div.input').hide()
  });
</script>
<form action="javascript:code_toggle()"><input type="submit" id="toggleButton" value="Show Code"></form>
```

```python
%matplotlib notebook
import time
import numpy as np
import pylab as pl

fig,ax = pl.subplots(1,1)

x,y = [], [] 

line, = ax.plot(x, y, 'ro')

for i in range(50):
    x.append(10*np.random.rand())
    y.append(np.exp(x[i]))
    line.set_data(x,y)
    ax.relim()
    ax.autoscale_view(True,True,True)
    fig.canvas.draw()
    time.sleep(0.3)
```
